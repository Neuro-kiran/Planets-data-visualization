```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
planets = sns.load_dataset('planets')
planets
```

```python
sns.histplot(planets['orbital_period'], bins=30, kde=True)
```

Insights:
    Most planets have short orbital periods, peaking at around 100 days.
    Fewer planets have longer orbits, creating a right-skewed distribution.

```python
sns.jointplot(x='orbital_period', y='mass', data=planets, kind='scatter')
```

Insights:
    There's no strong linear relationship between orbital period and mass.
    Planets are mainly concentrated in the lower mass and shorter orbital period region.

```python
sns.kdeplot(x=planets['orbital_period'], y=planets['mass'], cmap="Blues", fill=True)
```

```python
Insights:
    Confirms that most planets have low mass and short orbital periods.
    Shaded areas highlight common combinations of orbital period and mass.
```

```python
sns.pairplot(planets[['orbital_period', 'mass', 'year']], hue='year', diag_kind='kde')
```

Insights:
    No clear patterns suggest that the year of discovery significantly impacts these variables' relationships.
